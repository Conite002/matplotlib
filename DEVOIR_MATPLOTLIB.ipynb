```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.linspace(0,10,11)
```

```python
c = 3*10**8
E = m*(c**2)
E
```

```python
fig = plt.figure(num=1)
axe01 = fig.add_axes([0,0,1,1])
axe01.plot(m,E, color='r', lw=4)
axe01.set_xlabel("Masse en grammes")
axe01.set_ylabel("Energie en Joule")
axe01.set_title("E = mc^2")
```

```python
fig = plt.figure(num=1)
axe01 = fig.add_axes([0,0,1,1])
axe01.plot(m,np.log(E), color='r', lw=4)
axe01.set_xlabel("Masse en grammes")
axe01.set_ylabel("Energie en Joule")
axe01.set_title("E = mc^2")
axe01.grid(which="both", axis='y')
```

```python
july16_2007 = [4.75,4.98,5.08,5.01,4.89,4.89,4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']
labels_x = [int(str(x).split()[0]) for x in labels]
labels_x
```

```python
fig = plt.figure(num = 1)
axe02 = fig.add_axes([0,0,1,1])
axe02.spines['left'].set_color('blue')
axe02.spines['left'].set_linewidth(5)

axe02.plot(july16_2007,label="july16_2007", color='orange', lw = 2)
axe02.set_ylabel("2007 data", color='orange', fontsize= 18)
for ax in axe02.get_yticklabels():
    ax.set_color("orange")
axe02.legend()

axe03 = axe02.twinx()
axe03.plot(july16_2020,label="july16_2020", color='blue', lw = 2)
axe03.set_ylabel("2020 data", color= 'blue', fontsize = 18)
for ax in axe03.get_yticklabels():
    ax.set_color("blue")
axe03.spines['right'].set_color('blue')
axe03.spines['right'].set_linewidth(5)
axe03.legend()
```

```python
fig = plt.figure(num = 1)
axe02 = fig.add_axes([0,0,1,1])
axe02.plot(labels,july16_2007,label="july16_2007", color='orange', lw = 2)
axe02.plot(labels,july16_2020,label="july16_2020", color='blue', lw = 2)
axe02.legend(loc=[1.1,0.4])
```

```python
fig = plt.figure(num = 1)
axe02 = fig.add_axes([0,0,1,1])
axe02.spines['left'].set_color('orange')
axe02.spines['left'].set_linewidth(5)

axe02.spines['bottom'].set_color('red')
axe02.spines['bottom'].set_linewidth(5)

axe02.plot(july16_2007,label="july16_2007", color='orange', lw = 2)
axe02.set_ylabel("2007 data", color='orange', fontsize= 18)
for ax in axe02.get_yticklabels():
    ax.set_color("orange")
axe02.legend()

axe03 = axe02.twinx()
axe03.plot(july16_2020,label="july16_2020", color='blue', lw = 2)
axe03.set_ylabel("2020 data", color= 'blue', fontsize = 18)
for ax in axe03.get_yticklabels():
    ax.set_color("blue")
axe03.spines['right'].set_color('blue')
axe03.spines['right'].set_linewidth(5)
axe03.legend()
```

```python
fig = plt.figure(num = 1)
axe02 = fig.add_axes([0,0,1,1])
axe02.bar(labels,july16_2007,label="july16_2007", color='orange', lw = 2)
axe02.bar(labels,july16_2020,label="july16_2020", color='blue', lw = 2)
axe02.legend(loc=[1.1,0.4])
```

```python
fig = plt.figure(num = 1)
axe02 = fig.add_axes([0,0,1,1])
axe02.pie(july16_2007)
axe02.pie(july16_2020)
axe02.legend(loc=[1.1,0.4])
```
