```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = x*2
```

```python
plt.plot(x,y)
plt.title("droite y = 2x")
plt.xlabel("x_value")
plt.ylabel("y_value")
plt.xlim(0,20)
plt.ylim(0,20)
plt.savefig("fig001.png")
```

```python
x = np.linspace(0,10,10)
y = x **4

a = np.random.randint(1,100,20)
b = a*2
```

```python
fig = plt.figure(num=1,dpi= 200, facecolor = "#53EB34", edgecolor="#D13109")
axe01 = fig.add_axes([0,0,1,1])
axe02 = fig.add_axes([1.1,0.85,0.13,0.13])
```

```python
fig = plt.figure(num=1,dpi= 200, facecolor = "#C0D8BB", edgecolor="#D13109")
axe01 = fig.add_axes([0,0,1,1])
axe01.set_title("Graphe")
axe01.plot(x,y)
axe01.set_xlabel("X labels")
axe01.set_ylabel("Y labels")

axe02 = fig.add_axes([0.2,0.2,0.3,0.3])
axe02.set_xlim(3,6)
axe02.set_ylim(0,2000)
axe02.set_title("Zoom avant")
axe02.set_xlabel("a")
axe02.set_ylabel("b")
axe02.plot(x,y)

axe03 = fig.add_axes([-1,-1,1,1])
axe03.plot(x,y)

fig.savefig("figure02.png", bbox_inches='tight')
```

# Subplots 

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for axe in axes:
    axe.plot(x,y)

fig.subplots_adjust(hspace=2,wspace=.5)
fig.suptitle("Big title", fontsize=16, )

fig.savefig("figure03.png", bbox_inches= 'tight')
```
