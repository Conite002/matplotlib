```python
import matplotlib.pyplot  as plt
import numpy as np
```

```python
x = np.linspace(1,50,100)
```

```python
fig = plt.figure(num=1,dpi=100)
ax01 = fig.add_axes([0,0,1,1])
ax01.plot(x,x**4,label='X vs X**4',ls ="--", lw=3, color = "#ffff03")
line = ax01.plot(x,x,label="X vs 2X", ls=":", color = "#8ACE4A", lw=3)
ax01.legend(loc="upper right")
line[0].set_dashes([1,2,12])
```

```python
fig = plt.figure(num=1,dpi=100)
ax01 = fig.add_axes([0,0,1,1])
ax01.plot(x,x**2,label='X vs X**4',ls ="-", lw=1, color = "#ffff03", marker="s", markersize = 5)
line = ax01.plot(x,x+2,label="X vs 2X", ls=":", color = "#8ACE4A", lw=3)
ax01.legend(loc="upper right")
line[0].set_dashes([1,2,12])
```

```python
x = np.linspace(0,10,10)
fig = plt.figure(num=1, dpi=100)
axe03 = fig.add_axes([0,0,1,1])
line = axe03.plot(x,x**x,marker="^",label = "X vs X^2" , markersize=10, markeredgewidth = 2, markeredgecolor = "r")
axe03.plot(x,np.exp(x), label = "X vs exp(X)",marker="*", markersize=10, markeredgewidth = 2, markeredgecolor = "g")
axe03.plot(x,(np.log(x)), label = "X vs Sin(X)",marker="s", markersize=10, markeredgewidth = 2, markeredgecolor = "g")
axe03.legend()
axe03.set_ylim(0,500)
line[0].set_dashes([1,5,2,1])
#Zoom

axe04 = fig.add_axes([0.6,0.3,0.2,0.2])
axe04.plot(x,x**2 - x, ls = "--",marker="^",label = "X vs X^2 - X" , markersize=10, markeredgewidth = 2, markeredgecolor = "r")
axe04.plot(x,3*x**2-5*x, label = "X vs 3*X**2-5*X",marker="*", markersize=10, markeredgewidth = 2, markeredgecolor = "g")
axe04.plot(x,(np.log(x)), label = "X vs Sin(X)",marker="s", markersize=10, markeredgewidth = 2, markeredgecolor = "g")
axe04.set_xlim(6,10)
axe04.set_ylim(0,500)
axe04.grid(color='orange',alpha = 0.5,linewidth = 0.5, linestyle = "-")
```
